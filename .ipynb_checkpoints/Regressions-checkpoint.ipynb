## Data cleaning and initial EDA

```python
data = pd.read_csv('data/data_w_genres.csv')
corpus = data['genres']
corpus = corpus[corpus != '[]']
corpus = corpus.str.lstrip('[')
corpus = corpus.str.rstrip(']')
corpus = corpus.str.replace("'", '')
corpus = corpus.str.split(', ')
corpus.head()
```

There is a *huge* range of occurances, from 1 to over 600 occurances of a genre. A lot appear to have 25 occurances or fewer. It actually appears that almost half the data shows up only once or twice. I am somewhat arbitrarily going to pick anything below 25 occurances as a cutoff because I don't want to train my model on genres that aren't very prevalent.

```python
mask = frequency['occurances'] > 25
trim_freq = frequency[mask]
print("percent remaining: " + str(100*len(trim_freq)/len(frequency)))
print("length of dataset: "+ str(len(trim_freq)))
```

```python
trimmed_genre_sequences = []
for sequence in corpus:
    trimmed_genre_sequences.append([genre for genre in sequence if genre in list(trim_freq['genre'])])

# add this list to the data set first to not lose indexing
dataset = pd.DataFrame(corpus)
dataset['trimmed_genres'] = trimmed_genre_sequences

dataset.head(3)
```

```python
# remove empty lists to evaluate how many sequences are empty/removed from the data set.
trimmed_genre_sequences = [sequence for sequence in trimmed_genre_sequences if sequence != []]
trimmed_genre_sequences[:5]
```

```python
print(f"Data set size is now {len(trimmed_genre_sequences)} instead of original size of {len(corpus)}. This is {100*len(trimmed_genre_sequences)/len(corpus)}% of the original data")
```

Although over 80% of the genres were removed, 85% of the sequences still remain. This means I've been able to dramatically decrease the model's *labels* without severely impacting the data size. I find this acceptable and will clean up my dataset to removed any rows with empty lists in the `trimmed_genres` column.

```python
def boxplot_sorted(df, by, column, rot=0):
    # use dict comprehension to create new dataframe from the iterable groupby object
    # each group name becomes a column in the new dataframe
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    # find and sort the median values in this new dataframe
    meds = df2.median().sort_values()
    # use the columns in the dataframe, ordered sorted by median value
    # return axes so changes can be made outside the function
    return df2[meds.index].boxplot(rot=rot, return_type="axes", figsize=(8,80), vert=False)
```

```python
axes = boxplot_sorted(data_expanded, by = ['genre'], column = 'acousticness')
plt.ylabel('genre')
plt.title('acousticness')
plt.suptitle('')
plt.show()
```
